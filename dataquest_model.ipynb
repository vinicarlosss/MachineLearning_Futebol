```python

matches["date"] = pd.to_datetime(matches["date"])
matches["target"] = (matches["result"] == "W").astype("int")
matches["venue_code"] = matches["venue"].astype("category").cat.codes
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
matches['hour'] = matches['time'].fillna('N/A').str.replace(':.+', '', regex=True)
matches['hour'] = pd.to_numeric(matches['hour'], errors='coerce')
matches['hour'] = matches['hour'].fillna(matches['time'])
matches['hour'] = matches['hour'].replace({pd.NA: None}).astype('Int64')
matches["day_code"] = matches["date"].dt.dayofweek
matches
```

```python

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
train = matches[matches["date"] < '2023-12-10']
test = matches[(matches["date"] >= '2023-12-10') & (matches["date"] <= '2024-01-02')]
test.dropna(subset=['hour'], inplace=True)
test.dropna(subset=['result'], inplace=True)
predictors =["venue_code", "opp_code", "hour", "day_code"]
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
acc = accuracy_score(test["target"], preds)
acc
```

```python
combined = pd.DataFrame(dict(actual = test["target"], prediction = preds))
pd.crosstab(index=combined["actual"], columns=combined["prediction"])
```

```python
precision_score(test["target"], preds, average='micro')
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean().fillna(method='ffill').fillna(method='bfill')
    group[new_cols] = rolling_stats
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling.loc['Manchester City']
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
```

```python
matches_rolling.dropna(subset=['hour'], inplace=True)
matches_rolling
```

```python
def make_predictions(data, predictors):
    rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
    train = data[data["date"]<='2024-01-02']
    test = data[data["date"] > '2024-01-02']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual = test["target"], prediction = preds), index=test.index)
    predicao = pd.DataFrame(dict(prediction=preds), index=test.index)
    error = precision_score(test["target"], preds, average='micro')
    return combined, error, predicao
```

```python
combined, error, predicao = make_predictions(matches_rolling, predictors + new_cols)
```
